```python
df3['race_month'] = df3['race_date'].str.split('.').str.get(1).astype(int) 
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x : 'winter' if x > 12 else 'fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'winter') 
```
